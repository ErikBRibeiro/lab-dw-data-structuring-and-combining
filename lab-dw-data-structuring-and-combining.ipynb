```python
import pandas as pd
import numpy as np
import cleaning

pd.options.display.max_columns = None

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

df1 = pd.read_csv(url)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)
df1.head()
```

```python
df2.head()
```

```python
df3.head()
```

```python
def extract_middle_value(complaints):
    return int(complaints.split('/')[1])
```

```python
df1 = cleaning.standardize_column_names(df1)
df2 = cleaning.standardize_column_names(df2)
df3 = cleaning.standardize_column_names(df3)
df1 = df1.rename(columns={'st' : 'state'})
df2 = df2.rename(columns={'st' : 'state'})
df1 = df1.dropna(subset=['number_of_open_complaints'])
df2 = df2.dropna(subset=['number_of_open_complaints'])
df1['number_of_open_complaints'] = df1['number_of_open_complaints'].apply(extract_middle_value)
df2['number_of_open_complaints'] = df2['number_of_open_complaints'].apply(extract_middle_value)
# Had to turn customer_lifetime_value into float individually and befone combining the dataframes, because df3 was already cleaned and if i try to clean it together, all rows of df3 become NaN
cleaning.remove_percentage(df1, 'customer_lifetime_value')
cleaning.remove_percentage(df2, 'customer_lifetime_value')
df2 = df2[df1.columns]
df3 = df3[df1.columns]

df = pd.concat([df1, df2,df3], ignore_index=True)

df.head()
```

```python
# Here i wanted to fill null gender with random F or M, while mantaining the proportion it had already
def gender_fill(df):
    gender_distribution = df['gender'].value_counts(normalize=True)
    num_missing = df['gender'].isnull().sum()
    num_f_to_fill = int(gender_distribution['F'] * num_missing)
    num_m_to_fill = num_missing - num_f_to_fill
    fill_values = ['F'] * num_f_to_fill + ['M'] * num_m_to_fill

    np.random.shuffle(fill_values)

    df.loc[df['gender'].isnull(), 'gender'] = fill_values
```

```python
gender_fill(df)
df['gender'].unique()
```

```python
cleaning.fill_missing_values(df, 'customer_lifetime_value', method='mean')

df.isnull().sum()
```

```python
state_map = {
    "AZ": "Arizona",
    "Cali": "California",
    "WA": "Washington"
}

df['state'] = df['state'].replace(state_map)
df['state'].unique()
```

```python
df['gender'].unique()
```

```python
df['gender'] = df['gender'].str.replace('f(emale|emal|e)?', 'F', case=False, regex=True)
df['gender'] = df['gender'].str.replace('male|^m$', 'M', case=False, regex=True)

print(df['gender'].unique())
```

```python
df['education'].unique()
```

```python
education_map = {
    'Bachelors': 'Bachelor'
}

df['education'] = df['education'].replace(education_map)
df['education'].unique()
```

```python
df['policy_type'].unique()
```

```python
df['vehicle_class'].unique()
```

```python
vehicle_class_map = {
    'Sports Car': 'Luxury',
    'Luxury SUV': 'Luxury',
    'Luxury Car': 'Luxury'
}

df['vehicle_class'] = df['vehicle_class'].replace(vehicle_class_map)
df['vehicle_class'].unique()
```

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*
